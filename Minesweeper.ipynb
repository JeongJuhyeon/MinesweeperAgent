```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('Statistics copy.csv', index_col=0)
df['mine_density'] = [a/(b * b) for a, b in zip(df['no_mines'], df['gridsize'])]
print(df.head())
```

```python
sns.lmplot(x='mine_density', y='win_rate', data=df)
```
